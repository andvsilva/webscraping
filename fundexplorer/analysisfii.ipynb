```python
index = 0

for values in df_fiis['Preço Atual'].iteritems():
    
    string_ = str(values[1])
    
    if(len(string_) < 8):
        df_fiis.at[index, 'Preço Atual'] = float(string_)
        index = index + 1
        
    if(len(string_) == 8):
        df_fiis.at[index, 'Preço Atual'] = float(string_[0] + string_[2:] )
        index = index + 1
        
    if(len(string_) == 9):
        df_fiis.at[index, 'Preço Atual'] = float(string_[0:2] + string_[3:] )
        index = index + 1
```

```python
df_fiis["Preço Atual"] = df_fiis["Preço Atual"].astype('float64')
df_fiis.info()
```
