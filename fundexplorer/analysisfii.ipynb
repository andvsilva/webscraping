```python
import pandas as pd
```

```python
df_fiis = pd.read_csv('database/fiis.csv')
df_fiis.info()
```

```python
df_fiis['Preço Atual'] = df_fiis['Preço Atual'].str.replace('R$ ','', regex=False)
df_fiis['Preço Atual'] = df_fiis['Preço Atual'].str.replace(',','.', regex=False)
df_fiis['Dividendo'] = df_fiis['Dividendo'].str.replace('R$ ','', regex=False)
df_fiis['Dividendo'] = df_fiis['Dividendo'].str.replace(',','.', regex=False)
df_fiis['DividendYield'] = df_fiis['DividendYield'].str.replace('%','', regex=False)
df_fiis['DividendYield'] = df_fiis['DividendYield'].str.replace(',','.', regex=False)
df_fiis['DY (3M)Acumulado'] = df_fiis['DY (3M)Acumulado'].str.replace('%','', regex=False)
df_fiis['DY (3M)Acumulado'] = df_fiis['DY (3M)Acumulado'].str.replace(',','.', regex=False)
df_fiis['DY (6M)Acumulado'] = df_fiis['DY (6M)Acumulado'].str.replace('%','', regex=False)
df_fiis['DY (6M)Acumulado'] = df_fiis['DY (6M)Acumulado'].str.replace(',','.', regex=False)
df_fiis['DY (12M)Acumulado'] = df_fiis['DY (12M)Acumulado'].str.replace('%','', regex=False)
df_fiis['DY (12M)Acumulado'] = df_fiis['DY (12M)Acumulado'].str.replace(',','.', regex=False)

df_fiis['DY (3M)Média'] = df_fiis['DY (3M)Média'].str.replace('%','', regex=False)
df_fiis['DY (3M)Média'] = df_fiis['DY (3M)Média'].str.replace(',','.', regex=False)

df_fiis['DY (6M)Média'] = df_fiis['DY (6M)Média'].str.replace('%','', regex=False)
df_fiis['DY (6M)Média'] = df_fiis['DY (6M)Média'].str.replace(',','.', regex=False)

df_fiis['DY (12M)Média'] = df_fiis['DY (12M)Média'].str.replace('%','', regex=False)
df_fiis['DY (12M)Média'] = df_fiis['DY (12M)Média'].str.replace(',','.', regex=False)

df_fiis['DY Ano'] = df_fiis['DY Ano'].str.replace('%','', regex=False)
df_fiis['DY Ano'] = df_fiis['DY Ano'].str.replace(',','.', regex=False)

df_fiis['Variação Preço'] = df_fiis['Variação Preço'].str.replace('%','', regex=False)
df_fiis['Variação Preço'] = df_fiis['Variação Preço'].str.replace(',','.', regex=False)

df_fiis['Rentab.Período'] = df_fiis['Rentab.Período'].str.replace('%','', regex=False)
df_fiis['Rentab.Período'] = df_fiis['Rentab.Período'].str.replace(',','.', regex=False)

df_fiis['Rentab.Período'] = df_fiis['Rentab.Período'].str.replace('%','', regex=False)
df_fiis['Rentab.Período'] = df_fiis['Rentab.Período'].str.replace(',','.', regex=False)

df_fiis['Rentab.Acumulada'] = df_fiis['Rentab.Acumulada'].str.replace('%','', regex=False)
df_fiis['Rentab.Acumulada'] = df_fiis['Rentab.Acumulada'].str.replace(',','.', regex=False)

df_fiis['PatrimônioLíq.'] = df_fiis['PatrimônioLíq.'].str.replace('R$ ','', regex=False)

df_fiis['VPA'] = df_fiis['VPA'].str.replace(',','.', regex=False)
df_fiis['P/VPA'] = df_fiis['P/VPA'].str.replace(',','.', regex=False)

df_fiis['DYPatrimonial'] = df_fiis['DYPatrimonial'].str.replace('%','', regex=False)
df_fiis['DYPatrimonial'] = df_fiis['DYPatrimonial'].str.replace(',','.', regex=False)

df_fiis['VariaçãoPatrimonial'] = df_fiis['VariaçãoPatrimonial'].str.replace('%','', regex=False)
df_fiis['VariaçãoPatrimonial'] = df_fiis['VariaçãoPatrimonial'].str.replace(',','.', regex=False)

df_fiis['Rentab. Patr.no Período'] = df_fiis['Rentab. Patr.no Período'].str.replace('%','', regex=False)
df_fiis['Rentab. Patr.no Período'] = df_fiis['Rentab. Patr.no Período'].str.replace(',','.', regex=False)

df_fiis['Rentab. Patr.Acumulada'] = df_fiis['Rentab. Patr.Acumulada'].str.replace('%','', regex=False)
df_fiis['Rentab. Patr.Acumulada'] = df_fiis['Rentab. Patr.Acumulada'].str.replace(',','.', regex=False)

df_fiis['VacânciaFísica'] = df_fiis['VacânciaFísica'].str.replace('%','', regex=False)
df_fiis['VacânciaFísica'] = df_fiis['VacânciaFísica'].str.replace(',','.', regex=False)

df_fiis['VacânciaFinanceira'] = df_fiis['VacânciaFinanceira'].str.replace('%','', regex=False)
df_fiis['VacânciaFinanceira'] = df_fiis['VacânciaFinanceira'].str.replace(',','.', regex=False)

df_fiis['VPA'] = df_fiis['VPA'].str.replace('R$ ','', regex=False)
```

```python
index = 0

for values in df_fiis['Preço Atual'].iteritems():
    
    string_ = str(values[1])
    
    if(len(string_) < 8):
        df_fiis.at[index, 'Preço Atual'] = float(string_)
        index = index + 1
        
    if(len(string_) == 8):
        df_fiis.at[index, 'Preço Atual'] = float(string_[0] + string_[2:] )
        index = index + 1
        
    if(len(string_) == 9):
        df_fiis.at[index, 'Preço Atual'] = float(string_[0:2] + string_[3:] )
        index = index + 1
```

```python
index = 0

for values in df_fiis['PatrimônioLíq.'].iteritems():
    
    string_ = str(values[1])
    
    if(len(string_) == 12):
        df_fiis.at[index, 'PatrimônioLíq.'] = float((string_[0:1] + string_[2:5] + string_[6:]).replace(',','.'))
    
    if(len(string_) == 13):
        df_fiis.at[index, 'PatrimônioLíq.'] = float((string_[0:2] + string_[3:6] + string_[7:]).replace(',','.'))
        
    
    if(len(string_) == 14):
        string_to_float = string_[0:3] + string_[4:7] + string_[8:]
        string_to_float = string_to_float.replace(',','.')
        df_fiis.at[index, 'PatrimônioLíq.'] = float(string_to_float)
        #print(float(string_to_float))
    
    if(len(string_) == 16):
        string_to_float = string_[0:1] + string_[2:5] + string_[6:9] + string_[10:]
        string_to_float = string_to_float.replace(',','.')
        df_fiis.at[index, 'PatrimônioLíq.'] = float(string_to_float)
        #print(float(string_to_float))
    
    index = index + 1
```

```python
index = 0
for values in df_fiis['VPA'].iteritems():
    string_ = str(values[1])
    
    if(len(string_) == 8):
        df_fiis.at[index, 'VPA'] = float((string_[0:1] + string_[2:]))
        
    index = index + 1
```
